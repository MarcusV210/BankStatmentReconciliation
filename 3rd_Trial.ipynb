```python
columns = list(df.columns)
columns.remove('Customer ID')
columns.remove('Address')
columns.remove('Date Of Account Opening')
columns.remove('Last Transaction Date')
columns.remove('Loan ID')
columns.remove('Loan Amount')
columns.remove('Loan Type')
columns.remove('Interest Rate')
columns.remove('Loan Term')
columns.remove('Approval/Rejection Date')
columns.remove('Loan Status')
columns.remove('Credit Limit')
columns.remove('Credit Card Balance')
columns.remove('Minimum Payment Due')
columns.remove('Payment Due Date')
columns.remove('Last Credit Card Payment Date')
columns.remove('Rewards Points')
columns.remove('Feedback ID')
columns.remove('Feedback Date')
columns.remove('Feedback Type')
columns.remove('Resolution Status')
columns.remove('Resolution Date')
columns.remove('Anomaly')
```

```python
df = df[columns]
```

```python
df['Full name'] = df['First Name'] + ' ' + df['Last Name']
```

```python
df.drop(['First Name', 'Last Name'], axis=1, inplace=True)
```
